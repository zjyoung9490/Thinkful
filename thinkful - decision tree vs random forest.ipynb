# Readmittance Rate for Diabetes Treatment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
diabetic_data = pd.read_csv(r'\Users\mike1\Documents\Thinkful\diabetic_data.csv')

diabetes = diabetic_data.sample(n = 50000)
```

```python
# Make readmitted a binary classification and get dummies for categorical data
diabetes['readmitted'].replace(['>30', '<30'], 'YES', inplace = True)

# Readmitted will be the target variable
readmitted = pd.get_dummies(diabetes['readmitted'], drop_first = True)

diabetes.drop(['weight'], 1)
diabetes = pd.get_dummies(diabetes)
diabetes.head()
```

```python
# Set up for modelling
from sklearn import ensemble
from sklearn import tree
from sklearn.model_selection import cross_val_score

X = diabetes.drop('readmitted_NO', 1)
Y = readmitted
```

```python
# Decision Tree Model
decision_tree = tree.DecisionTreeClassifier(criterion = 'entropy', max_features = 1, max_depth = 5)
dtc = decision_tree.fit(diabetes, readmitted)

cross_val_score(dtc, X, Y, cv=10)

#note: changing max_depth not affecting accuracy
```

```python
# Random Forest Model
rfc = ensemble.RandomForestClassifier()

cross_val_score(rfc, X, Y, cv=10)
```

Comparing the decision tree with the random forest model, the accuracy of the random forest is significantly greater. After cross validation all tests had greater than 96% accuracy in the random forest model. The decision tree model on the other hand never had an accuracy above 55%. Changing the max depth of the decision tree did not have any noticeable affect on the accuracy. The run time for the decision tree was much shorter however, as would be expected. The random forest took over a minute to run for a dataset consisting of only 50,000 rows.
